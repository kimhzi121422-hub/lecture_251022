# 11. 누적 막대 그래프
---
### 누적 막대 그래프의 활용
- 누적 막대 그래프는 여러 데이터 계열을 하나의 막대에 쌓아서 표현한다.
- 전체 합계와 각 구성 요소의 비율을 동시에 파악할 수 있다.
- 항목별 비교와 전체 합계를 한눈에 볼 수 있어 유용하다.


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 사용자
matplotlib.rcParams['font.size'] = 15 # 폰트 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
```

### 간단한 예제 데이터
- 먼저 간단한 예제로 누적 막대 그래프의 원리를 이해해보자.


```python
# 간단한 예제 데이터
students = ['박태근', '이영희', '박지성']
korean = [85, 90, 78]
english = [88, 85, 92]
math = [90, 88, 85]
```

## 누적 막대 그래프 만들기
### bottom 파라미터
- **bottom** 파라미터를 사용하여 막대를 쌓을 수 있다.
- 첫 번째 막대 위에 두 번째 막대를 쌓으려면 bottom에 첫 번째 데이터를 전달한다.


```python
#국어점수만 표시
plt.bar(students, korean)
plt.bar(students, english, bottom=korean)

plt.title("국어점수")
plt.ylabel('점수')
```

```python
import matplotlib.pyplot as plt

cities = ['서울', '대전', '대구', '부산']
population = [970, 342, 242, 295]
colors = ['red', 'blue', 'green', 'orange']
hatch = ['/', '\\', 'x', 'O']  # 각막대 무늬(패턴) 정의

fig, ax = plt.subplots()

bar = ax.bar(cities, population, color=colors)

ax.set_title('도시별 인구')
ax.set_ylabel('인구(만명)')
ax.set_ylim(200, 1000)

for idx, rect in enumerate(bar):
    # 막대 높이
    height = rect.get_height()
    # x좌표+폭/2(막대 중앙)
    ax.text(rect.get_x() + rect.get_width()/2, height + 30, f'{population[idx]}만명',
            ha='center', va='bottom', color=colors[idx], fontsize=12, fontweight='bold')
    bar[idx].set_hatch(hatch[idx])

plt.tight_layout()
plt.show()
```

```python
plt.barh(cities,population)
```

```python
#국어점수만 표시
plt.bar(students, korean, label='국어')
plt.bar(students, english, bottom=korean, label='영어')
# math는 누적된 국어+영어 위에 쌓임
bottom_math = [korean[i]+english[i] for i in range(len(students))]
plt.bar(students, math, bottom=bottom_math, label='수학')

plt.title("국어점수+영어점수+수학점수")
plt.ylabel('점수')
```

```python
import numpy as np

plt.bar(students, korean, label='국어')
plt.bar(students, english, bottom=korean, label='영어')
plt.bar(students, math, bottom=np.array(korean)+np.array(english), label='수학')

plt.title("국어점수+영어점수+수학점수")
plt.ylabel('점수')
```

### 세 개 이상 데이터 누적하기
- 세 번째 막대를 쌓으려면 이전 두 데이터의 합을 bottom에 전달한다.


### np.array
- 여러개의 값을 담을 수 있는 다차원 그리드
- 1차원(배열), 2차원(행렬), 3차원 이상(텐서)의 형태를 가질 수 있다.

**특징**
- 안의 모든 요소가 동일한 자료형이어야한다.
- 요소별 연산이 가능하다. (배열전체에 대한 한번의 연산으로)
